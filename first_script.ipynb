```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.remote.webelement import WebElement
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import time
import csv
```

```python
#Test Script For getting links from a page and saving in csv file
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://www.yell.com/s/cake+makers+and+decorations-doncaster.html')
flag = True
while flag:
    try:
        with open('bussinessLinks.csv', 'a+', encoding='utf-8-sig', newline='') as f:
            writer = csv.writer(f)
            flag2 = True
            while flag2:
                try:
                    brandLinks = WebDriverWait(driver, 5).until(
                                                    EC.presence_of_all_elements_located((By.XPATH, "//a[@class='businessCapsule--title']")) )

                    for i in brandLinks:
                        link = i.get_attribute('href')
                        print(link)
                        writer.writerow([link])
                        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                    nextBtn = WebDriverWait(driver, 5).until(
                                                    EC.presence_of_element_located((By.XPATH, " //a[@class='btn btn-blue btn-fullWidth pagination--next']")) )
                    time.sleep(2)
                    nextBtn.click()
                except:
                    print('break1')
                    flag2 =False
                break

                    
    except:
        print('break2')
        flag = False
        break
```

```python
#now search for each brand link from csv and getting required data gather in a new csv
with open("bussinessLinks.csv") as f:
    data = csv.reader(f)
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.maximize_window()
    for i in data:
        for j in i:
            
#             link = j[3:]
            print(j)
            try:
                driver.get(j)
                name = WebDriverWait(driver, 5).until(
                                                    EC.presence_of_element_located((By.XPATH, "//h1[@class='text-h1 businessCard--businessName']")) )
                number = WebDriverWait(driver, 5).until(
                                                    EC.presence_of_element_located((By.XPATH, "//span[@class='business--telephoneNumber']")) )
                website = WebDriverWait(driver, 5).until(
                                                    EC.presence_of_element_located((By.XPATH, "//a[@class='btn btn-big btn-yellow businessCard--callToAction']")) )
                address = WebDriverWait(driver, 5).until(
                                                    EC.presence_of_element_located((By.XPATH, "//span[@class='address']")) )
                with open('sampleData3.csv', 'a+', encoding='utf-8-sig', newline='') as f:
                    writer = csv.writer(f)
                    writer.writerow([name.text,number.text,website.get_attribute('href'),address.text])
                print(name.text)
                print(number.text)
                print(website.get_attribute('href'))
                print(address.text)
            except:
                print('break1')
                try:
                    name = WebDriverWait(driver, 5).until(
                                                        EC.presence_of_element_located((By.XPATH, "//h1[@class='text-h1 businessCard--businessName']")) )
                    number = WebDriverWait(driver, 5).until(
                                                        EC.presence_of_element_located((By.XPATH, "//span[@class='business--telephoneNumber']")) )

                    address = WebDriverWait(driver, 5).until(
                                                        EC.presence_of_element_located((By.XPATH, "//span[@class='address']")) )
                    with open('sampleData3.csv', 'a+', encoding='utf-8-sig', newline='') as f:
                        writer = csv.writer(f)
                        writer.writerow([name.text,number.text,'none',address.text])
                    print(name.text)
                    print(number.text)
                    print(address.text)
                except:
                    print('break 2')
                    try:
                        name = WebDriverWait(driver, 5).until(
                                                            EC.presence_of_element_located((By.XPATH, "//h1[@class='text-h1 businessCard--businessName']")) )
                        number = WebDriverWait(driver, 5).until(
                                                            EC.presence_of_element_located((By.XPATH, "//span[@class='business--telephoneNumber']")) )

                        with open('sampleData3.csv', 'a+', encoding='utf-8-sig', newline='') as f:
                            writer = csv.writer(f)
                            writer.writerow([name.text,number.text,'none','none'])
                        print(name.text)
                        print(number.text)
                        print(address.text)
                    except:
                        print('break 3')
                
            
                
```
