```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.remote.webelement import WebElement
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import time
import csv
```

```python
with open("towns.csv") as f:
    data = csv.reader(f)
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.maximize_window()
    for i in data:
        for j in i:
            print(j)
            driver.get(f"https://www.yell.com/s/cake+makers+and+decorations-{j}.html")
            with open('bhundLinks.csv', 'a+', encoding='utf-8-sig', newline='') as f:
                writer = csv.writer(f)
                try:
                    for i in range(10):
                        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                        nextBtn = WebDriverWait(driver, 10).until(
                                                        EC.presence_of_element_located((By.XPATH, " //a[@class='btn btn-blue btn-fullWidth pagination--next']")) )
                        writer.writerow([nextBtn.get_attribute('href')])
                        print(nextBtn.get_attribute('href'))
                        nextBtn.click()
                        time.sleep(5)
                except:
                    print('break')
                    

```
